### Problem 15: Triangle Classifier Function

**Description:**
Create a function that classifies a triangle based on its three sides.

**What you need to do:**
1. Create function `classify_triangle(a, b, c)` that takes 3 sides
2. Return a STRING:
   - `"Invalid"` if it's not a valid triangle
   - `"Equilateral"` if all sides are equal
   - `"Isosceles"` if exactly two sides are equal
   - `"Scalene"` if all sides are different

**Triangle validity rule:** Sum of any two sides must be greater than the third side

**Example usage:**
```python
result = classify_triangle(5, 5, 5)
print(result)  # Output: Equilateral

result = classify_triangle(5, 5, 8)
print(result)  # Output: Isosceles

result = classify_triangle(3, 4, 5)
print(result)  # Output: Scalene

result = classify_triangle(1, 2, 10)
print(result)  # Output: Invalid
```

````python
class TriangleData:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

class TriangleClassifier(TriangleData): # Inheritance
    def classify_triangle(self):
        # 1. Validity Rule: Sum of any two sides > third side
        if (self.a + self.b <= self.c) or (self.a + self.c <= self.b) or (self.b + self.c <= self.a):
            return "Invalid"

        # 2. Check for Equilateral (All sides equal)
        if self.a == self.b == self.c:
            return "Equilateral"

        # 3. Check for Isosceles (Exactly two sides equal)
        elif self.a == self.b or self.b == self.c or self.a == self.c:
            return "Isosceles"

        # 4. Otherwise Scalene (All sides different)
        else:
            return "Scalene"

# Testing the logic
def test_classifier(a, b, c):
    obj = TriangleClassifier(a, b, c)
    print(obj.classify_triangle())

# Example usage
test_classifier(5, 5, 5)   # Output: Equilateral
test_classifier(5, 5, 8)   # Output: Isosceles
test_classifier(3, 4, 5)   # Output: Scalene
test_classifier(1, 2, 10)  # Output: Invalid
````

# Because 5! = 5 × 4 × 3 × 2 × 1 = 120

### Problem 19: Password Validator Function

**Description:**
Create a function that checks if a password is strong and prints the result.

**What you need to do:**
1. Create function `validate_password(password)`
2. Check if password has:
   - At least 8 characters
   - At least one digit (0-9)
   - At least one uppercase letter (A-Z)
   - At least one lowercase letter (a-z)
3. If strong: print "Strong password!"
4. If weak: print which requirements are missing

**Example usage:**
```python
validate_password("Pass1234")
# Output: Strong password!

validate_password("hello")
# Output: Weak password:
# - Too short (needs at least 8 characters)
# - No digit
# - No uppercase letter
```

**Hint:** Use `.isdigit()`, `.isupper()`, `.islower()` to check characters

````python
class PasswordData:
    def __init__(self, password):
        self.password = password

class PasswordValidator(PasswordData): # Inheritance
    def validate_password(self):
        missing = []

        # 1. Check length
        if len(self.password) < 8:
            missing.append("Too short (needs at least 8 characters)")

        # Flags for other requirements
        has_digit = False
        has_upper = False
        has_lower = False

        # Loop through each character to check requirements
        for char in self.password:
            if char.isdigit():
                has_digit = True
            if char.isupper():
                has_upper = True
            if char.islower():
                has_lower = True

        # Add missing requirements to the list
        if not has_digit:
            missing.append("No digit")
        if not has_upper:
            missing.append("No uppercase letter")
        if not has_lower:
            missing.append("No lowercase letter")

        # Final Output logic
        if not missing:
            print("Strong password!")
        else:
            print("Weak password:")
            for reason in missing:
                print(f"- {reason}")

# Test Cases
validator1 = PasswordValidator("Pass1234")
validator1.validate_password()

print("\n---") # Separator

validator2 = PasswordValidator("hello")
validator2.validate_password()
````

````python
# Test your function
validate_password("Pass1234")
validate_password("hello")
validate_password("100%Strong")
````

### Problem 20: Discount Calculator Function

**Description:**
Create a function that calculates the final price after applying a discount percentage.

**What you need to do:**
1. Create function `calculate_discount(price, discount_percent)`
2. Calculate the discount amount: `price × (discount_percent / 100)`
3. Calculate final price: `price - discount_amount`
4. Return the final price rounded to 2 decimal places

**Example 1:**
```python
final_price = calculate_discount(1000, 10)
print(final_price)
# Output: 900.0
```

**Example 2:**
```python
final_price = calculate_discount(500, 25)
print(final_price)
# Output: 375.0
```

**Example 3:**
```python
final_price = calculate_discount(1299, 15)
print(final_price)
# Output: 1104.15
```

**Note:**
- To calculate X% of a number: `number × (X / 100)`
- Example: 10% of 1000 = 1000 × (10/100) = 1000 × 0.1 = 100

````python
class Pricing_Data:
    def __init__(self, price, discount_percent):
        self.price = price
        self.discount_percent = discount_percent

class DiscountCalculator(Pricing_Data): # Inheritance
    def calculate_discount(self):
        # 1. Calculate discount amount: price * (discount_percent / 100)
        discount_amount = self.price * (self.discount_percent / 100)

        # 2. Calculate final price: price - discount_amount
        final_price = self.price - discount_amount

        # 3. Return rounded result to 2 decimal places
        return round(final_price, 2)

# Test cases using the class logic
def get_final_price(p, d):
    obj = DiscountCalculator(p, d)
    return obj.calculate_discount()

print(get_final_price(1000, 10))
print(get_final_price(500, 25))
print(get_final_price(1299, 15))
````
