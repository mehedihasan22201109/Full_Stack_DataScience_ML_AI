# Variables

A variable is like a **labeled container** that stores data. Compares to other programming languages, Python has no command for declaring a variable. A variable is created the moment you first assign a value to it.

```python
name = "Ayan"  # Variable 'name' stores a string
age = 7         # Variable 'age' stores a number
```
---
## Variable Naming Rules
1. Can only contain letters, numbers, and underscores
2. Must start with a letter or underscore (`_`)
3. Case-sensitive (`name` and `Name` are different)
4. Cannot use Python keywords (if, for, while, etc.)
---
### Valid Names:
```python
student_name = "Ayan"
_age = 20
subject1 = "Math"
qwerty = "random"
```

---
### Invalid Names:
```python
2students = "Error"  # Starts with number
student-name = "Error"  # Contains hyphen
for = "Error"  # Python keyword
```


---
## Multi-Word Variable Names

When a variable name has multiple words, there are different naming conventions:
```python
# Snake Case

# Camel Case
studentName = "Ayan"

# Pascal Case

```

# Python Input/Output (I/O)

Programs need to **interact** with users:
- Take data from the user (**Input**)
- Process/work on that data
- Show results back to the user (**Output**)

## Where Does This Happen? The Console!

The **console** (also called terminal or command prompt) is the place where:
- **You type** → **Program receives** (INPUT: Console → Program)
- **Program displays** → **You read** (OUTPUT: Program → Console)

Think of it like a conversation:
- **Input** = You talking to the program
- **Output** = Program talking back to you

# Python Strings
Strings are sequences of characters enclosed in quotes.

## Creating Strings
```python
# Single or double quotes (both work the same)
name = 'Ayan'
message = "Hello, World!"

# Multi-line strings (use triple quotes - either ''' or """)
long_text = """This is a
multi-line
string"""

# This also works
another_text = '''This is also
a multi-line
string'''
```
## Common String Methods

````python
text = "  hello world  "

# Case conversion
print(text.upper())       # '  HELLO WORLD  '
print(text.lower())       # '  hello world  '
print(text.title())       # '  Hello World  '
````

````python
# Removing spaces
print(text.strip())       # 'hello world' (removes spaces from both ends)
print(text.lstrip())      # 'hello world  ' (removes left spaces)
print(text.rstrip())      # '  hello world' (removes right spaces)
````

````python
# Replace
print(text.replace('hello', 'hi'))  # '  hi world  '
````

````python
# Checking content
print('hi' in text)                      # False
print(text.strip().startswith('hello'))  # True
print(text.strip().endswith('world'))    # True
````

````python
# Count and Find
print(text.count('l'))    # 3 (how many times 'l' appears)
print(text.find('world')) # 8 (position where 'world' starts)
````

# Conditional Statements

Conditional statements allow your program to make decisions and execute different code based on conditions.

## The `if` Statement

**Syntax:**
```
if condition:
    code to execute
```


````python
age = 27
# age = int(input())
if age >= 18:
    print("You are an adult")
    print("You can vote")
````

## The `if-else` Statement

**Syntax:**
```
if condition:
    code to execute if True
else:
    code to execute if False
```

````python
age = 15
# age = int(input())
if age >= 18:
    print("You are an adult")
else:
    print("You are a minor")
````

## The `if-elif-else` Statement

**Syntax:**
```
if condition1:
    code to execute if condition1 is True
elif condition2:
    code to execute if condition2 is True
elif condition3:
    code to execute if condition3 is True
else:
    code to execute if all conditions are False
```

````python
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")
````

## Logical Operators

Combine multiple conditions:

### `and` - Both conditions must be True


````python
age = 20
has_id = True

if age >= 18 and has_id:
    print("You can enter")
else:
    print("Entry denied")
````

### `or` - At least one condition must be True

````python
day = "Sunday"

if day == "Sunday" or day == "Wednesday":
    print("We have class today!")
else:
    print("No class today")
````

### `not` - Reverses the condition

````python
is_raining = False

if not is_raining:
    print("Let's go outside!")
else:
    print("Stay inside")
````

# For Loop

A `for` loop is used to iterate over a sequence (like a list, string, or range) and execute code for each item.

**Syntax:**
```
for variable in sequence:
    code to execute
```

````python
word = "Python"

for char in word:
    print(char)
````

````python
fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    print(fruit)
````

````python
for i in range(5):
    print(i)
````

````python
for i in range(10, 0, -1):
    print(i, end=" ")
````

````python
word = "Python"

for index, char in enumerate(word):
    print(f"{index}: {char}")
````

````python
# Example: Count vowels in a string
text = "Hello World"
vowels = "aeiouAEIOU"
count = 0

for char in text:
    if char in vowels:
        count += 1

print(f"Number of vowels: {count}")
````

# `break` Statement

Exits the loop immediately, even if the loop condition is still `True`.

**Syntax:**
```
for variable in sequence:
    if condition:
        break
    code to execute
```

````python
for i in range(10):
    if i == 5:
        break
    print(i)
````

---
# `continue` Statement

Skips the rest of the current iteration and moves to the next one.

**Syntax:**
```
for variable in sequence:
    if condition:
        continue
    code to execute
```

````python
# Skip even numbers
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
````

---
# `pass` Statement

Does nothing. Used as a placeholder when a statement is required syntactically.

**Syntax:**

```
    if condition:
        pass
    code to execute
```

````python
for i in range(10):
    if i == 3:
        pass
    
    if i == 5:
        continue
    
    if i == 8:
        break
    
    print(i)
````

# Functions

A function is a reusable block of code that performs a specific task. Functions help organize code and avoid repetition.

**Syntax:**
```
def function_name():
    code to execute
```

````python
def greet():
    print("Hello, World!")

greet()
greet()
````

## Function with Parameters

Parameters allow you to pass data into functions.

**Syntax:**
```
def function_name(parameter1, parameter2):
    code to execute
```

````python
def greet(name):
    print(f"Hello, {name}!")

greet("Ayan")

greet("Kawchar")
````

````python
def introduce(name, age, city):
    print(f"My name is {name}, I am {age} years old, and I live in {city}.")

introduce("Ayan", 15, "Dhaka")
````

## Function with Return Value

`return` sends a value back to the caller.

**Syntax:**
```
def function_name(parameters):
    code to execute
    return value
```

````python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)

print(f"Total: {add(10, 20)}")
````

## Function with Default Parameters

**Syntax:**
```
def function_name(param1, param2=default_value):
    code to execute
```

````python
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet("Ayan")

greet("Ayan", "Hi")

greet("Ayan", "Good morning")
````

## Variable Scope: Local vs Global
Variables created **inside** a function are **local** - only accessible inside that function. <br>
Variables created **outside** functions are **global** - accessible everywhere.


````python
def my_function():
    z = 10  # Local variable
    print(f"Inside function: {z}")

my_function()

# print(z)
# Error: NameError: name 'z' is not defined
# z doesn't exist outside the function!
````

````python
z = 10  # Global variable

def my_function():
    print(f"Inside function: {z}")

my_function()

print(f"Outside function: {z}")
````

### Local Variable Shadows Global Variable
If a local variable has the same name as a global variable, the local one is used inside the function.

````python
x = 10  # Global variable

def my_function():
    x = 20  # Local variable (different from global x)
    print(f"Inside function: {x}")

my_function()

print(f"Outside function: {x}")
````

### Modifying Global Variables Inside Functions

Use the `global` keyword to modify a global variable inside a function.

````python
count = 0

def increment():
    count = count + 1  # ❌ Error!
    # Python thinks you're creating a local variable
    # but trying to use it before assignment

# increment()
````

````python
count = 0  # Global variable

def increment():
    global count
    count += 1
    print(f"Inside function: {count}")

print(f"Before: {count}")

increment()

increment()

print(f"After: {count}")
````
